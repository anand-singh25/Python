```python
# Importing Library

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Loading Dataset

epl_df = pd.read_csv("D:\Courses\Portfolio\Excel Projects\EPL\EPL 2020-21 Data Set.csv")
epl_df.head()
```

```python
 epl_df.info()
```

```python
epl_df.describe()
```

```python
epl_df.isna().sum()
```

```python
# Creating new columns

epl_df['Minutes_Per_Match'] = epl_df['Mins'] / epl_df['Matches'].astype(int)
epl_df['Goals_Per_Match'] = epl_df['Goals'] / epl_df['Matches'].astype(float)
epl_df.head()
```

```python
# Totals Goals

total_goals = epl_df['Goals'].sum()
total_goals
```

```python
# Total Penalty Goals

total_penalty_goals = epl_df['Penalty_Goals'].sum()
total_penalty_goals
```

```python
# Total Penalty Attempted

total_penalty_attempted = epl_df['Penalty_Attempted'].sum()
total_penalty_attempted
```

```python
# Pie chart for Penalties missed VS scored

plt.figure(figsize=(14,7))
total_penalty_not_scored = total_penalty_attempted - total_penalty_goals
data = [total_penalty_not_scored , total_penalty_goals]
labels = ['Penalties Scored' , 'Penalties Missed']
color = sns.color_palette()
plt.pie(data, labels = labels , colors = color , autopct = '%0.f%%')
plt.title('Penalties Scored VS Missed')
plt.show()
```

```python
# Bar Chart for Most Players from which country 

nationality = epl_df.groupby('Nationality').size().sort_values(ascending=False)
nationality.head(10).plot(kind = 'bar' , figsize = (12,6))
```

```python
# Clubs with Maximum players in their squad

# epl_df['Club'].value_counts().head(6)

epl_df['Club'].value_counts().nlargest(6)
```

```python
# Bar chart to show Clubs with Maximum players in their squad

max_players = epl_df['Club'].value_counts().nlargest(5)
max_players.plot(kind = 'bar' , figsize = (12,5))
```

```python
# Clubs with Minium players in their squad

# epl_df['Club'].value_counts().head(6)

epl_df['Club'].value_counts().nsmallest(6)
```

```python
# Bar chart to show Clubs with Minimum players in their squad

min_players = epl_df['Club'].value_counts().nsmallest(5)
min_players.plot(kind = 'bar' , figsize = (12,7))
plt.title('Clubs with Minimum Players')
```

```python
# Players based on Age Group

under_20 = epl_df[epl_df['Age'] <= 20]
age_20_25 = epl_df[(epl_df['Age'] > 20) & (epl_df['Age'] <= 25)]
age_25_35 = epl_df[(epl_df['Age'] > 25) & (epl_df['Age'] <= 30)]
above_30 = epl_df[epl_df['Age'] > 30]
```

```python
age_group = pd.array([under_20['Name'].count(), age_20_25['Name'].count(),
                     age_25_35['Name'].count(), above_30['Name'].count()])
labels = ['<=20' , '> 20 & <= 25', '> 25 & <= 30' , '>30']
plt.title('Total Players with Age')
plt.pie(age_group, labels = labels , autopct = '%0.1f%%')
plt.show()
```

```python
# Total players under age 20 in each club

under_20['Club'].value_counts().plot(kind = 'bar')
```

```python
# Under age 20 in ManU

under_20[under_20['Club'] == "Manchester United"]
```

```python
# Average age of players in each club

plt.figure(figsize=(12,7))
sns.boxplot(x = 'Club' , y = 'Age' , data = epl_df)
plt.xticks(rotation = 75)
```

```python
num_players = epl_df.groupby('Club').size()
data = epl_df.groupby('Club')['Age'].sum() / num_players
data.sort_values(ascending=False)
```

```python
# Total assists from each club

assists_by_clubs = pd.DataFrame(epl_df.groupby('Club', as_index=False)['Assists'].sum())
ax = sns.barplot(x = 'Club', y = 'Assists', data = assists_by_clubs.sort_values(by = 'Assists',ascending=False))
ax.set_xlabel('Club')
ax.set_ylabel('Assists')
plt.xticks(rotation = 75)
plt.title('Total assists from each club')
plt.show()
```

```python
# Total Goals from each club

goals_by_clubs = epl_df.groupby('Club', as_index=False)['Goals'].sum()
ax = sns.barplot(x = 'Club', y = 'Goals', data = goals_by_clubs.sort_values(by = 'Goals', ascending=False)[:10])
ax.set_xlabel('Club')
ax.set_ylabel('Goals')
plt.xticks(rotation = 75)
plt.title('Total Goals from each club')
plt.show()
```

```python
# Most goals by players

most_goals = epl_df[['Name', 'Club', 'Goals', 'Matches']].nlargest(10, columns = 'Goals')
most_goals
```

```python
# Least goals by players

least_goals = epl_df[['Name', 'Club', 'Goals', 'Matches']].nsmallest(10, columns = 'Goals')
least_goals
```

```python
# Goals per match
# Goals_Per_Match is a column which we created earlier and passing now

goals_per_match = epl_df[['Name', 'Goals_Per_Match', "Goals", 'Matches']].nlargest(10, columns = 'Goals_Per_Match')
goals_per_match
```

```python
# Pie Chart for Goals with and without assists

plt.figure(figsize=(14,7))
assists = epl_df['Assists'].sum()
data = [total_goals - assists , assists]
labels = ["Goals Without Assists", "Goals With Assists"]
plt.pie(data , labels = labels , autopct = '%0.1f%%')
plt.title('Goals with and without assists', fontsize = 10)
plt.show()
```

```python
# Top 10 players with most yellow cards

yellow_cards = pd.DataFrame(epl_df.groupby('Name', as_index=False)['Yellow_Cards'].sum())
ax = sns.barplot(x = 'Name', y = 'Yellow_Cards', data = yellow_cards.sort_values(by = 'Yellow_Cards', ascending=False)[:10])
ax.set_xlabel('Name')
ax.set_ylabel('Yellow Cards')
plt.xticks(rotation = 75)
plt.title('Total Yellow Cards')
plt.show()
```

```python
# alternative code for above section

total_yellow_cards = epl_df.sort_values(by = 'Yellow_Cards', ascending=False)[:10]
ax = sns.barplot(x = 'Name', y= 'Yellow_Cards', data = total_yellow_cards)
ax.set_xlabel('Name')
ax.set_ylabel('Number of Yellow Cards')
plt.xticks(rotation= 75)
plt.title('Total Yellow Cards')
plt.show()
```
